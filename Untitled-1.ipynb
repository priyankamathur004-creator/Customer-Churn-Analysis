```python
#Handle missing values- here in the given data set the Churn Date has some missing data
for i in CustomerOtt_df.columns:
    if CustomerOtt_df[i].dtype == 'object':
        m = CustomerOtt_df[i].mode()[0]
        CustomerOtt_df[i] = CustomerOtt_df[i].fillna(m)
    else:
        me = CustomerOtt_df[i].mean()
        CustomerOtt_df[i]= CustomerOtt_df[i].fillna(me)
```

3. Behavioural Analysis 
 Compare:  
o Engagement vs churn  
o Subscription type vs retention  
o Upgrade offered vs accepted  

```python
# Engagement vs churn 
CustomerOtt_df['Engagement_by_churn'] = (
    CustomerOtt_df['Engagement_Score'] / CustomerOtt_df['Churn_Rate']
)
CustomerOtt_df['Engagement_by_churn']
```

```python
#Subscription type vs retention
CustomerOtt_df['Retention'] = CustomerOtt_df['Status'].apply(
lambda x: 'Retained'if ['Status']=='Active'else 'Churned'
)
```

```python
Retention_by_type = (
    CustomerOtt_df
    .groupby(['SubscriptionType','Retention'])['CustomerID']
    .count()
    .reset_index()
    .rename(columns={'CustomerID':'Count'})
)
Retention_by_type
```

```python
# Upgrade offered vs accepted  
CustomerOtt_df[['UpgradeOffered' ,'UpgradeAccepted']]
```

```python
#Upgrade offered vs accepted  
Upgrade_offered = (CustomerOtt_df['UpgradeOffered'] == 'Yes').sum()
Upgrade_accepted = (CustomerOtt_df['UpgradeAccepted'] == 'Yes').sum()

Upgrade_acceptance_rate = Upgrade_accepted / Upgrade_offered * 100
print("Upgrade Acceptance Rate:", Upgrade_acceptance_rate)
```

```python
pd.crosstab(CustomerOtt_df['UpgradeOffered'], CustomerOtt_df['UpgradeAccepted'])
```

```python
valid_data = CustomerOtt_df[
    ~((CustomerOtt_df['UpgradeAccepted'] == 'Yes') & (CustomerOtt_df['UpgradeOffered'] == 'No'))
]

Upgrade_offered = (valid_data['UpgradeOffered'] == 'Yes').sum()
Upgrade_accepted = (valid_data['UpgradeAccepted'] == 'Yes').sum()

Upgrade_acceptance_rate = (Upgrade_accepted / Upgrade_offered) * 100
print("Valid Upgrade Acceptance Rate:", Upgrade_acceptance_rate)
```

```python
#• Group customers by JoinDate  
Customers_by_joindate = CustomerOtt_df.groupby(by='JoinDate')['CustomerID'].count().sort_values(ascending= False)
Customers_by_joindate = Customers_by_joindate.reset_index()
Customers_by_joindate

```

```python
Retention_overtime = (
    CustomerOtt_df
    .groupby(CustomerOtt_df['JoinDate'].dt.to_period('M'))['Retention']
    .value_counts()
    .unstack(fill_value=0)
    .reset_index()
    .rename(columns={'JoinDate':'Month'})
)

print(Retention_overtime)
```

```python
# Cohort retention heatmap

Retention_matrix = CustomerOtt_df.pivot_table(

    index='SubscriptionType',
    columns=CustomerOtt_df['JoinDate'].dt.to_period('M'),
    values='Retention',
    aggfunc='count'
)

plt.figure(figsize=(12,6))
sns.heatmap(Retention_matrix, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title("Retention Over Time by Subscription Type")
plt.show()
```

```python
#Revenue by segment  
Revenue_by_Segment = CustomerOtt_df.groupby(by = 'Customer_segmentation')['RevenueGenerated'].sum().sort_values(ascending=False).reset_index()
Revenue_by_Segment
```

```python
median_revenue = CustomerOtt_df['RevenueGenerated'].median()

High_value_customers = CustomerOtt_df[
    (CustomerOtt_df['RevenueGenerated'] >= median_revenue) &
    (CustomerOtt_df['Retention'] == 'Retained')
]

print(High_value_customers[['CustomerID','RevenueGenerated','SubscriptionType']])
```

```python
#At Risk customers

avg_watch = CustomerOtt_df['WatchHoursPerWeek'].mean()

At_risk_users = CustomerOtt_df[
    (CustomerOtt_df['WatchHoursPerWeek'] < avg_watch) |
    (CustomerOtt_df['Retention'] == 'Churned')
]

print(At_risk_users[['CustomerID','WatchHoursPerWeek','Retention','SubscriptionType']])
```
